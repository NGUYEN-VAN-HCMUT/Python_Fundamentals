2. Lập trình kiểm tra một số nguyên `x` bất kỳ có nằm trong khoảng từ 10 đến 20 hay không?
*   Nếu `x` nằm trong khoảng đó: In ra "Nằm trong khoảng [10, 20]" \
*   Nếu `x` không nằm trong khoảng đó: In ra "Không nằm trong khoảng [10, 20]"


```
# Lập trình tại đây
x = float(input('nhập vào một số x bất kỳ'))
if x >= 10 and x <=20:
  print('Nằm trong khoảng [10, 20]')
else:
  print('Không nằm trong khoảng [10, 20]')

```

3. Một công ty chỉ tuyển dụng nhân sự dưới hoặc bằng 50 tuổi và ở Hà Nội, lập trình mệnh đề điều kiện kiểm tra biến `age` xem nhỏ hơn hoặc lớn hơn 50 và `at_hanoi` là `True` hay `False` :



*   Nếu `age` nhỏ hơn hoặc bằng 50 và `at_hanoi` là `True`: In ra "Hoàn toàn đủ điều kiện"
*   Các trường hợp còn lại: In ra "Không đủ điều kiện"





```
# Lập trình tại đây
#age = 50
#at_hanoi = True 

age = int(input('Nhập vào số tuổi'))
at_hanoi = True

if age <= 50 and at_hanoi == True:
  print('Hoàn toàn đủ điều kiện')
else:
  print('Không đủ điều kiện')
```
